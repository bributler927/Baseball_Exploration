```python
import pandas as pd
import numpy as np
```

```python
people = pd.read_csv("baseballdatabank-2022.2/core/People.csv")
salaries = pd.read_csv("baseballdatabank-2022.2/contrib/Salaries.csv")
pitching_stats = pd.read_csv("baseballdatabank-2022.2/core/Pitching.csv")
appearences = pd.read_csv("baseballdatabank-2022.2/core/Appearances.csv")
```

```python
'''
This code chunk obtains the ids of all pitchers in the statcast era of the MLB
by comparing their number of pitching appearances vs their appearances in other positions.
The logic I used in this is that a pitcher has made more appearances pitching than
every other position.
'''
#narrow down appearances by statcast era (2015 to now)
appearences = appearences[appearences.yearID >= 2015]

#all available positions other than defined subsitution positions
positions = ["G_p","G_c","G_1b","G_2b","G_3b","G_ss","G_lf","G_cf","G_rf","G_of"]

#same positions without pitching
other_positions = ["G_c","G_1b","G_2b","G_3b","G_ss","G_lf","G_cf","G_rf","G_of"]
p_ids = appearences.groupby("playerID")[positions].sum()

#array to save the pitcher ids for future crosslisting
pitcher_ids = []

#choose a pitcher based on their number of non-pitching appearances
for index, player in p_ids.iterrows():
    other_total = 0
    for pos in other_positions:
        other_total += player[pos]
    
    if player["G_p"] > other_total:
        pitcher_ids.append(index)

pitcher_ids
```
